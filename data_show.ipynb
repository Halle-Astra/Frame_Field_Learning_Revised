```python
from matplotlib import pyplot as plt 
import numpy as np 
import json 
from pycocotools.coco import COCO
import os 
%matplotlib inline 
```

```python
root = '/data/sda/datasets/Datasets4FFL/mapping_challenge_dataset/raw/val'
froot = os.path.join(root,'annotation.json')

coco = COCO(froot)
```

```python
img_ls = coco.loadImgs(ids=[0])
```

```python
imgname = img_ls[0]['file_name']
```

```python
help(coco.getAnnIds)
```

```python
ann_ids = coco.getAnnIds(imgIds = [0])
```

```python
ann_ls = coco.loadAnns(ids = ann_ids)
```

```python
img = plt.imread(os.path.join(root,'images',imgname))
plt.imshow(img)

for ann_dict in ann_ls:
    ann = ann_dict['segmentation']
    ann = np.array(ann).reshape(-1,2)
    plt.plot(ann[:,0],ann[:,1],'b')
    
    bbox = ann_dict['bbox']
    y,x,y1,x1 = bbox
    bbox_xyxy = [[x,y],[x1,y1]]
    bbox_xyxy = np.array(bbox_xyxy).reshape(-1,2)
    
    plt.imshow(img)
    
    plt.plot(bbox_xyxy[:,0],bbox_xyxy[:,1],'r-*')
    plt.title(bbox)
    plt.show()
```
